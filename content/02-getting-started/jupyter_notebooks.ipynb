```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_gradient(x):
    return sigmoid(x) * (1 - sigmoid(x))
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([[1]])
Y = np.array([[1]]).T

np.random.seed(1)
weights = np.random.random((X.shape[1], 1))
bias = np.random.random((1, 1))

display(weights, bias)
```

```python
output = X * weights + bias
pred = sigmoid(output)
pred
```

```python
(Y-pred)**2
```

Back propagation

```python
dsigma = 2 * (pred - Y)
db = sigmoid_gradient(output) * dsigma
# Усе інше одинички
dW1 = db * X
dW1
```

```python
bias -= dW1
weights -= dW1
```

```python
pred = sigmoid(np.dot(X, weights) + bias)
pred
```

Те саме через keras

```python
from keras.layers import Dense
from keras.models import Sequential
import tensorflow as tf

model = Sequential([
    Dense(1, input_shape=(1,), activation='sigmoid')
])

model.summary()
```

```python
X_toy = tf.constant([1])
Y_toy = tf.constant([1])
```

Градієнтний спуск

```python

optimizer = tf.keras.optimizers.SGD(learning_rate=1.0)
loss_fn = tf.keras.losses.MSE
```

```python
params = [np.empty(shape = (1,1), dtype = np.float32), np.empty(shape = (1,), dtype = np.float32)]
params[0][0] = 0.417
params[1][0] = 0.72

model.set_weights(params)
model.get_weights()
```

```python
for epoch in range(100):
    with tf.GradientTape() as tape:
      # прямий прохід
        pred = model(X_toy)
        print('Prediction is', pred.numpy())

        loss_value = loss_fn(Y_toy, pred)
        print('Loss is', loss_value.numpy())

        # Зворотній прохід
        grads = tape.gradient(loss_value, model.trainable_weights)
        print('Grad are', grads)

    optimizer.apply_gradients(zip(grads, model.trainable_weights))
```

```python
from keras.layers import Dense
from keras.models import Sequential
import tensorflow as tf

tf.random.set_seed(1)
model = Sequential([
    Dense(1, input_shape=(1,), activation='sigmoid')
])

model.summary()
```

```python
for epoch in range(1):
    with tf.GradientTape() as tape:
        pred = model(X_toy)
        print('Prediction is', pred.numpy())

        loss_value = loss_fn(y_toy, pred)
        print('Loss is', loss_value.numpy())

        grads = tape.gradient(loss_value, model.trainable_weights)
        print('Grad are', grads)

    optimizer.apply_gradients(zip(grads, model.trainable_weights))
```
